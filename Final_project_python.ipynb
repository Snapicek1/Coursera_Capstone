```python
df_day1 = df_incidents[['Day', 'ID']]

df_day1
```

```python
from pandas.api.types import CategoricalDtype
cats = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
cat_type = CategoricalDtype(categories=cats, ordered=True)
df_day1['Day'] = df_day1['Day'].astype(cat_type)
df_day1 = df_day1.groupby(['Day']).count()
df_day1
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
# group countries by continents and apply sum() function 
# note: the output of the groupby method is a `groupby' object. 
# we can not use it further until we apply a function (eg .sum())

#df_continents.head()
# step 2: plot data
df_day1.plot(kind='bar', 
             figsize=(10, 6)
            )

plt.xlabel('Day of week') # add to x-label to the plot
plt.ylabel('Number of accidents') # add y-label to the plot
plt.title('Number of car accidents during the week') # add title to the plot

plt.show()
```

```python

df_date = df_incidents[['Datum', 'ID']].groupby(['Datum']).count()

df_date
```

```python
df_date.plot(kind='bar', 
             figsize=(20, 5),
             label=None
            )

plt.xlabel('Number of accidents per day') # add to x-label to the plot
plt.ylabel('Day') # add y-label to the plot
plt.title('Number of accidents during the year') # add title to the plot

plt.show()
```

```python
df_root = df_incidents[['MainCause','ID']]

df_root = df_root.groupby('MainCause', axis=0).count()
df_root
```

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
df_root.plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', # add in percentages
                            startangle=45,     # start angle 90° (Africa)
                            shadow=True,       # add shadow  
                            pctdistance=1.12,  
                            subplots=True,
                            labels=None
                                        )

plt.title('Car accident by main caused')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.legend(labels=df_root.index, loc= 'upper left')
plt.show()
```
